```python
def fibonacci(n):
    if n <= 1:
        return n 
    return fibonacci(n - 1) + fibonacci(n-2)
```

```python
tmp = fibonacci(10)
print(tmp)
```

```python
def lastDigitFibonacciNumber(n):
    if n <= 1:
        return n 
    else:
        last_number = fibonacci(n)
        return last_number%10
```

```python
tmp = lastDigitFibonacciNumber(331)
print(tmp)
```
